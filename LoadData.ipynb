```python
import pandas as pd
import os
import shutil
```

```python
root_dir = "extracted"
labeled_images = []
visited = set()

for genre in os.listdir(root_dir): 
    folder_path = os.path.join(root_dir, genre)

    
    if not os.path.isdir(folder_path): 
        continue
        
    print(folder_path)
    for img in os.listdir(folder_path): 
        if img in visited: 
            print('hit visited')
            continue
        visited.add(img)
        img_path = os.path.join(folder_path, img)
        labeled_images.append({'img_path': img_path, 'label': genre})
    
    
paintings_labeled = pd.DataFrame(labeled_images)
paintings_labeled.to_csv('WikiArts_labeled.csv', index=False)
        
```
